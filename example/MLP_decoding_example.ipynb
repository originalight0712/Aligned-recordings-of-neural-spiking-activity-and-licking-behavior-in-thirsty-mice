```python
import torch
import scipy.io
import os
import scipy.io
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import accuracy_score
```

```python
#data preparation
def prepare_data(X, Y):
    """
    Prepare the data: convert data to the format for decoding
    """
    X_processed = []
    y_processed = []
    
    for i in range(len(X)):
        # Assume each array in X has shape (features, time_points)
        # Transpose to make time points as rows and features as columns
        x_data = X[i]['array']  # Shape becomes (20000, 32)
        
        # Assume each array in Y has shape (time_points, 1) or (time_points,)
        y_data = Y[i]['array'].flatten()  # Ensure it's a 1D array
        
        # Check if lengths match
        if len(x_data) != len(y_data):
            print(f"Warning: Length mismatch in sample {i}: {len(x_data)} vs {len(y_data)}")
            min_len = min(len(x_data), len(y_data))
            x_data = x_data[:min_len]
            y_data = y_data[:min_len]
        
        X_processed.append(x_data)
        y_processed.append(y_data)
    
    return X_processed, y_processed
```

```python
# build the model

# Check GPU and set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define MLP model
class BinaryMLP(nn.Module):
    def __init__(self, input_size=32, hidden_size1=100, hidden_size2=50):
        super(BinaryMLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.fc3 = nn.Linear(hidden_size2, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x

# Initialize model and move to GPU
model = BinaryMLP().to(device)
criterion = nn.BCELoss()  # Binary cross entropy loss
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
spike,trial = prepare_data(spike_list, trial_list)
```

```python
# train the model
print("Preparing training data...")
all_X_train = []
all_y_train = []
X_train_list = spike
y_train_list = trial

for i in range(len(X_train_list)):
    X_train = X_train_list[i]  # Transposed to 20000×32
    y_train = y_train_list[i].flatten()
    
    if len(X_train) != len(y_train):
        min_len = min(len(X_train), len(y_train))
        X_train = X_train[:min_len]
        y_train = y_train[:min_len]
    
    all_X_train.append(X_train)
    all_y_train.append(y_train)

# Combine all training data
X_train_combined = np.vstack(all_X_train)
y_train_combined = np.hstack(all_y_train)

# Convert to PyTorch tensors
X_tensor = torch.FloatTensor(X_train_combined).to(device)
y_tensor = torch.FloatTensor(y_train_combined).unsqueeze(1).to(device)  # Add dimension

# Create data loader
dataset = TensorDataset(X_tensor, y_tensor)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)

# 2. Training loop (standard deep learning approach)
num_epochs = 10
print(f"\nStarting training, total {num_epochs} epochs...")

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    
    for batch_x, batch_y in dataloader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    # Print statistics for each epoch
    avg_loss = epoch_loss / len(dataloader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Average loss: {avg_loss:.6f}")

print("Training completed")

# 3. Test the model
test_accuracies = []
print("\nStarting testing...")

for i in range(len(X_test_list)):
    # Prepare test data
    X_test = X_test_list[i]
    y_test = y_test_list[i].flatten()
    
    if len(X_test) != len(y_test):
        min_len = min(len(X_test), len(y_test))
        X_test = X_test[:min_len]
        y_test = y_test[:min_len]
    
    # Convert to tensor
    X_tensor = torch.FloatTensor(X_test).to(device)
    
    # Predict
    model.eval()
    with torch.no_grad():
        outputs = model(X_tensor)
        predictions = (outputs.cpu().numpy() >= 0.5).astype(int).flatten()
    
    accuracy = accuracy_score(y_test, predictions)
    test_accuracies.append(accuracy)
    print(f"Test sample {i+1}: Accuracy = {accuracy:.4f}")

# 4. Analyze results
average_accuracy = np.mean(test_accuracies)
print(f"\nFinal average test accuracy: {average_accuracy:.4f}")
print(f"Accuracy standard deviation: {np.std(test_accuracies):.4f}")
print(f"Minimum accuracy: {np.min(test_accuracies):.4f}")
print(f"Maximum accuracy: {np.max(test_accuracies):.4f}")

# 5. Save the model
torch.save(model.state_dict(), 'trained_mlp_model.pth')
print("Model saved as trained_mlp_model.pth")
```
